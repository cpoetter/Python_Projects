# FVTK displayed inline in iPython Notebook Example

FVTK is a 3D renderer for python. Like probably most other python programmers I work regulary with iPython Notebooks. Unfortunately iPython does not support inline display option for VTK. With the script found here https://pyscience.wordpress.com/2014/09/03/ipython-notebook-vtk/, however, it does.

This small example should illustrate this little helper function for plotting VTK and FVTK inside an iPython Notebook. 

First we have to import some packages to create some data for the actual plot.

```python
from dipy.viz import fvtk
from dipy.data import get_sphere
import numpy as np
```

This example shows a net of spheres genereted in the following block.

```python
r = fvtk.ren()
odfs = np.ones((5, 5, 724))
odfs[..., 0] = 2.
sphere = get_sphere('symmetric724')
actor = fvtk.sphere_funcs(odfs, sphere)
actor.RotateY(35)
actor.RotateZ(45)
fvtk.add(r, actor)
```

Now the important part: We want to avoid the VTK internal function fvtk.show(r), which will automatically open an external GUI.

Insead we import our own little package.

```python
from utilities import vtk_show
```

The next step is to execute the function with the VTK randoring as parameter.

```python
vtk_show(r)
```

The function also takes optional parameter for the width and height. The predefined values are both 750px.

```python
vtk_show(r, 500, 400)
```
